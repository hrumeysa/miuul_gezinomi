```python
# Şehir-Concept kırılımında PRICE ortalamalarınedir?

data.groupby(by=["SaleCityName","ConceptName"]).agg({"Price" : "mean"}).reset_index()
```

```python
# SaleCheckInDayDiff değişkenini kategorik bir değişkene çeviriniz

bins = [-1,7,30,90, data["SaleCheckInDayDiff"].max()] # değerlerinin hangi aralıklara ayrılacağını belirlendi 
labels = ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]
# her aralığın hangi kategoriye ait olduğunu açıklayan etiketler

data["EB_Score"] = pd.cut(data["SaleCheckInDayDiff"], bins, labels=labels)
# SaleCheckInDayDiff değerlerini bu aralıklara ve etiketlere göre kategorik bir sütuna dönüştürdük 
data.head(50).to_excel("eb_score.xlsx", index=False) # dönüştürülmüş ver bir Excel dosyasına kaydedildi
```

```python
Şehir-Concept-EB Score, Şehir-Concept- Sezon, Şehir-Concept-CInDay kırılımında ortalama ödenen ücret ve 
yapılan işlem sayısı cinsinden inceleyiniz ?
```

```python
# Şehir-Concept-EB_Score krılımında ücret ortalamaları

data.groupby(by=["SaleCityName", "ConceptName", "EB_Score"]).agg({"Price" : ["mean","count"]}).reset_index()
```

```python
# Şehir-Concept-Sezon krılımında ücret ortalamaları

data.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price" : ["mean","count"]}).reset_index()
```

```python
# Şehir-Concept-CInDay krılımında ücret ortalamaları

data.groupby(by=["SaleCityName", "ConceptName", "CInDay"]).agg({"Price" : ["mean","count"]}).reset_index()
```

```python
# City-Concept-Season kırılımının çıktısını PRICE'a göre sıralayınız

agg_data = data.groupby(by=["SaleCityName", "ConceptName", "CInDay"]).agg({"Price" : "mean"}).reset_index().sort_values("Price", ascending =False)
agg_data.head(20)
```

```python
# Yeni seviye tabanlı müşterileri (persona) tanımlayınız.
# Yeni eklenecek değişkenin adı: sales_level_based

agg_data["sales_level_based"] = agg_data[["SaleCityName", "ConceptName", "CInDay"]].agg(lambda x: '_'.join(x).upper(), axis=1)
```

```python
# Yeni müşterileri (personaları) segmentlere ayırınız.

agg_data["SEGMENT"] = pd.cut(agg_data["Price"], 4, labels=["D", "C", "B", "A"])
agg_data.head(20)
```

```python
agg_data.groupby("SEGMENT").agg({"Price" : ["mean", "max", "sum"]}).reset_index()
```

```python
# Yeni gelen müşterileri sınıflandırıp, ne kadar gelir getirebileceklerini tahmin ediniz.

agg_data.sort_values(by="Price")
```
